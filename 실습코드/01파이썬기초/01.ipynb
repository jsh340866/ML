```python
score = 82
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
else:
    grade = 'C'
print(f'{score}점 → {grade}')
```

```python
scores = [90, 85, 77]
for s in scores:                 # 리스트 순회
    print('점수:', s)
total = 0
for s in scores:
    total = total + s
print('합계:', total, '/ 평균:', total / len(scores))
for i in range(3):               # 0,1,2
    print('반복', i)
```

```python
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    return 'C'
print(get_grade(95))
print(get_grade(83))
print(get_grade(70))
```

```python
scores = [90, 85, 77, 60]
passed = [s for s in scores if s >= 80]   # 80 이상만
doubled = [s * 2 for s in scores]         # 각 값 2배
print('합격:', passed)
print('2배:', doubled)
```

```python
# 1) 연산자 -- 조건문/필터의 재료
print(7 // 3, 7 % 3, 2 ** 3, 7 / 3)              # 몫, 나머지, 거듭제곱
print(3 == 3, 3 != 5, 3 < 5, 5 >= 5)      # 비교 → True/False
print(True and False, True or False, not True)   # 논리 and/or/not
print(90 in [90, 85], 'age' in {'age': 25})      # 멤버십 in

# 2) while -- 조건이 참인 동안 반복
n = 3
while n > 0:
    print(n)
    n = n - 1        # 조건을 바꿔야 멈춘다 (안 그러면 무한 반복)

# 3) 튜플 -- 값 묶음(수정 불가). 함수 다중 반환이 이 튜플
point = (3, 5)
x, y = point               # 언패킹
print(x, y)

# 4) 중첩 자료구조 -- 리스트 안 딕셔너리 = DataFrame의 재료
rows = [{'이름': '홍길동', '점수': 90}, {'이름': '김철수', '점수': 85}]
print(rows[0]['이름'])      # 홍길동

# 5) 딕셔너리 안전 접근 & None
d = {'age': 25}
print(d.get('city'))              # 없는 키 → None (에러 안 남)
print(d.get('city', '서울'))       # 없으면 기본값 '서울'
print(None is None, bool(0), bool(''))   # None 확인 / 0·빈문자 = False
```

```python
# 11-1. 딕셔너리 순회 -- for k, v in d.items() (전처리 반복에서 자주)
info = {'age': 25, 'city': '서울'}
for k, v in info.items():
    print(k, '->', v)

# 11-2. enumerate / zip -- 번호와 함께 / 두 리스트를 짝지어
names = ['A', 'B', 'C']; scores = [90, 85, 77]
for i, name in enumerate(names):    print(i, name)      # 0 A / 1 B / 2 C
for name, s in zip(names, scores):  print(name, s)      # A 90 / ...

# 11-3. sorted + lambda -- 기준을 정해 정렬 (특성 중요도 등)
pairs = [('Fare', 0.25), ('Age', 0.11), ('Sex', 0.40)]
print(sorted(pairs, key=lambda x: x[1], reverse=True))  # 중요도 내림차순

# 11-4. 문자열 메서드 -- 데이터 정리
print('  Hello, ML  '.strip())      # 양끝 공백 제거
print('a,b,c'.split(','))           # ['a','b','c']
print('2025년'.replace('년', ''))    # '2025'

# 11-5. f-string 포맷 -- 평가 출력(소수 자리 / 퍼센트)
acc = 0.85283
print(f'정확도: {acc:.4f}')          # 0.8528
print(f'정확도: {acc:.1%}')          # 85.3%
```
